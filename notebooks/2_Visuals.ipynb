Data Loadaing

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from src.config.paths import  RAW_DATA , VISUAL
```

```python
df = pd.read_csv(RAW_DATA)
sns.set_theme(style='darkgrid')
```

# Age

```python
sns.histplot(data=df , x='age')
plt.title("Age Frequency")
plt.savefig(VISUAL/'age/age-freq.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
sns.lineplot(data=df,x='age',y='bmi',hue='sex')
plt.title("BMI To Age")
plt.savefig(VISUAL/'age/age-bmi.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
sns.boxplot(data=df,x='age')
plt.title("Medians")
plt.savefig(VISUAL/'age/age-boxplot.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
sns.lineplot(data=df,x='age',y='charges',hue='sex')
plt.title("Charges To Age")
plt.savefig(VISUAL/'age-charges.png',dpi=300,bbox_inches='tight')
plt.show()
```
